OrdinalEncoder - Normalizar usando ordem

```python
size_df = pd.DataFrame(
    {
        "name" : ['Fred', 'John', 'Matt'],
        "size" : ['small', 'med', 'xxl']

    }
)
```

```python
ore = ce.OrdinalEncoder(
    mapping=[
        {
            "col" : "size",
            "mapping" : {
                "small" : 1,
                "med" : 2,
                "lg" : 3       
            }
        }
    ]
)

ore.fit_transform(size_df)
```

```python
#Mais de uma coluna
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['small','med','xxl']])
oe.fit_transform(size_df[['size']])
```

## **Make Column Transformer**
Mais de uma transformação de uma vez. Ela é útil quando há necessidade de realizar diferentes tipos de pré-processamento em diferentes colunas do conjunto de dados, como normalização em uma coluna e encoding em outra.

```python
teste = pd.read_csv('http://bit.ly/kaggletrain', nrows=6)
cols = ['Fare', 'Embarked', 'Sex', 'Age']
teste_X = teste[cols]
teste_X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

ohe = OneHotEncoder()
imp = SimpleImputer()
```

```python
from sklearn.compose import make_column_transformer

ct = make_column_transformer(
    (ohe, ['Embarked', 'Sex']),  # apply OneHotEncoder to Embarked and Sex
    (imp, ['Age']),              # apply SimpleImputer to Age
    remainder='passthrough')     # include remaining column (Fare) in the output

ct.fit_transform(teste_X)
pd.DataFrame(_)
```
