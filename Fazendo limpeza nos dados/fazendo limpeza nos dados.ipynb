## **Nome de colunas**


Ao usar pandas, ter nomees de colunas apropriadas torna possível o acesso aos atribuots

```python
import janitor as jn
Xbad = pd.DataFrame(
    {
        "A" : [1, None, 3],
        "sales number" : [20.0, 30.0, None]
    }
)
Xbad
```

Transforma colunas em letras minúsculas e os espaços substituidos por _

Recomendado atualizar as colunas usando atribuições com índices, o método .assign ou .loc e .iloc

Não é recomendado usar atribuições para atualizar as colunas no pandas (risco de sobrescrever métodos existentes com o mesmo nome de uma coluna)

```python
jn.clean_names(Xbad)
```

```python
def clean_col(name):
  return (
      name.strip().lower().replace(" ", "_")
  )

Xbad.rename(columns=clean_col)
```
